```python
linklist=[]
limit = 0
while limit<2000:
    Job_in_India='https://www.linkedin.com/jobs/search/?currentJobId=3402061428&geoId=102713980&keywords=job%20in%20india&location=India&refresh=true'+str(limit)
    linklist.append(Job_in_India)
    limit+=25
```

```python
len(linklist)
```

```python
link = []
for i in linklist:
    browser.get(i)
    time.sleep(10)
    links = browser.find_elements('tag name','a')
    for i in links:
        link.append(i.get_attribute('href'))
```

```python
href=[]
for i in link:
    try:
        if 'jobs/view' in i:
            href.append(i)
    except:
        pass
```

```python
href_final=set(href)
len(href_final)
```

```python
href_final = list(href_final)
```

```python
Designation = []
Location= []
Company_Name=[]
LinkedIn_Followers=[]
Emp_Details=[]
Inv_Level=[]
Applicants=[]

for i in range(0,500):
    browser.get(href_final[i])
    time.sleep(20)
    try:
        j = browser.find_element("class name","jobs-unified-top-card__job-title")
        Designation.append(j.text)
    except:
        Designation.append('NA')
    try:
        l = browser.find_element("class name","jobs-unified-top-card__bullet")
        Location.append(l.text)
    except:
        Location.append('NA')
    try:
        c = browser.find_element("class name","jobs-unified-top-card__company-name")
        Company_Name.append(c.text)
    except:
        Company_Name.append('NA')
    try:
        i = browser.find_elements(By.CLASS_NAME,"jobs-unified-top-card__job-insight")
        Inv_Level.append(i[0].text)
        Emp_Details.append(i[1].text)
    except:
        Inv_Level.append('NA')
        Emp_Details.append('NA')
    try:
        f = browser.find_element("class name","artdeco-entity-lockup__subtitle")
        LinkedIn_Followers.append(f.text)
    except:
        LinkedIn_Followers.append('NA')
    try:
        ap = browser.find_element("class name","jobs-unified-top-card__applicant-count")
        Applicants.append(ap.text)
    except:
        Applicants.append('NA')
```
